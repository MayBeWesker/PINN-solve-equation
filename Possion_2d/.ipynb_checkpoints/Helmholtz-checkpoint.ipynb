## PINN主干网络

```python
class MLP(nn.Module):
    def __init__(self, mlp_layers):
        super(MLP, self).__init__()
        self.model = nn.Sequential()
        for i in range(len(mlp_layers)-2):
            self.model.add_module(f'fc{i+1}', nn.Linear(mlp_layers[i], mlp_layers[i+1], bias=True))
            self.model.add_module(f'act{i+1}', nn.Tanh())
        self.model.add_module(f'fc{len(mlp_layers)-1})', nn.Linear(mlp_layers[-2], mlp_layers[-1], bias=False))
    
    def forward(self, X):
        return self.model(X)
    
    
mlp = MLP(mlp_layers)
mlp_param = MLP(mlp_layers_param)
mlp.model
```

## PINN模型

```python
def grad(outputs, inputs):
    return torch.autograd.grad(outputs, inputs,
                               grad_outputs=torch.ones_like(outputs),
                               create_graph=True)
```

```python
class PINN(nn.Module):
    def __init__(self, backbone, nn_param, mu=None, sigma=None):
        super(PINN, self).__init__()
        self.backbone = backbone
        self.nn_param = nn_param
        
        self.k = k
        if mu is not None and sigma is not None:
            self.is_inputs_normalization = True
            self.mu = mu
            self.sigma = sigma
            print(f'forward with normalization, mu={self.mu.tolist()}, sigma={self.sigma.tolist()}')
        else:
            self.is_inputs_normalization = False
            
        self.m = temperature.mean()
        self.s = temperature.max() - temperature.min()
    
    def forward(self, X_res, f_res, T_data):
        f_res_pred = self.net_f(X_res)
        T_pred = self.net_u(X_res)

        loss_res = torch.mean((f_res_pred - f_res) ** 2)
        loss_data = torch.mean((T_pred - T_data) ** 2)
        return loss_res, loss_data
    
    def net_u(self, X):
        if self.is_inputs_normalization == True:
            X = (X - self.mu) / self.sigma
        return self.backbone(X) * self.s + self.m
    
    def net_param(self, X):
        if self.is_inputs_normalization == True:
            X = (X - self.mu) / self.sigma
        return self.nn_param(X)
    
    def net_f(self, X):
        X.requires_grad_(True)
        T = self.net_u(X)
        
        # 求实部u的梯度
        grad_u = grad(T, X)[0]
        u_x = grad_u[:, [0]]
        u_y = grad_u[:, [1]]
        u_z = grad_u[:, [2]]
        
        ku_xx = grad(self.k * u_x, X)[0][:, [0]]
        ku_yy = grad(self.k * u_y, X)[0][:, [1]]
        ku_zz = grad(self.k * u_z, X)[0][:, [2]]
        
        G = self.net_param(X)
        
        f_res_pred = ku_xx + ku_yy + ku_zz + G
        
        return f_res_pred
    

pinn = PINN(mlp, mlp_param)
```

## Adam训练

```python
X_res, f_res, T_data
```

```python
# 训练
dataset = Dataset(domain)
X_res, f_res, T_data = dataset.train_data()
X_res = torch.from_numpy(X_res).float().to(device)
f_res = torch.from_numpy(f_res).float().to(device)
T_data = torch.from_numpy(T_data).float().to(device)

mu = X_res.mean(dim=0)
sigma = X_res.std(dim=0)  # 求样本标准差

backbone = MLP(mlp_layers)
nn_param = MLP(mlp_layers_param)
pinn = PINN(backbone, nn_param, mu, sigma)
pinn = pinn.to(device)

optimizer_adam = optim.Adam(pinn.backbone.parameters(), lr=learning_rate)
optimizer_adam_param = optim.Adam(pinn.nn_param.parameters(), lr=learning_rate)
# optimizer, mode='min', factor=0.1, patience=10, threshold=0.0001, 
# threshold_mode='rel', cooldown=0, min_lr=0, eps=1e-08, verbose=False
lr_sche = ReduceLROnPlateau(optimizer_adam, mode='min', factor=gamma, verbose=True)
lr_sche_param = ReduceLROnPlateau(optimizer_adam_param, mode='min', factor=gamma, verbose=True)

logger = {
    "loss": [],
    "loss_res": [],
    "loss_data": [],
    "iter": [],
    "mu": mu,
    "sigma": sigma
}
with open(train_info_path + 'train_info.txt', 'w') as f:
    f.write('Training by Adam:\n')
start_time = time.time()

for it in range(n_iters):
    pinn.zero_grad()
    loss_res, loss_data = pinn(X_res, f_res, T_data)
    loss = 0*loss_res + 1000*loss_data
    loss.backward()
    optimizer_adam.step()
    optimizer_adam_param.step()
    
    # 保存训练信息 -------
    logger["loss"].append(loss.item())
    logger["loss_res"].append(loss_res.item())
    logger["loss_data"].append(loss_data.item())
    logger["iter"].append(it + 1)
    
    if (it + 1) % 100 == 0:  # 打印训练信息 -------
        info = f'Iter # {it+1:4d}/{n_iters}\ttime:{time.time()-start_time:.1f}\t' + f'loss:{loss.item():.2e}, loss_r:{loss_res.item():.2e}, loss_d:{loss_data.item():.2e}'
        with open(train_info_path + 'train_info.txt', 'a') as f:
            f.write(info + '\n')
        print(info)
        
    if (it + 1) % 100 == 0:  # 学习率衰减 -------
        lr_sche.step(loss)
        lr_sche_param.step(loss)
```
