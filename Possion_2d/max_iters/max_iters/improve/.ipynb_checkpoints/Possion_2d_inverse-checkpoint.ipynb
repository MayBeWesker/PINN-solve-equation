```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import ReduceLROnPlateau

import numpy as np
import matplotlib.pyplot as plt
import scipy.io as io
import pandas as pd

import os
import time
```

## 超参数设置

```python
domain = (0, 0.25, 0, 0.25)
xmin, xmax, ymin, ymax = domain
mlp_layers = [2] + [60]*8 + [1]
mlp_layers_G = [2] + [20]*2 + [1]
n_iters = 20000
# n_iters = 200000
learning_rate = 1e-3
gamma = 0.6
k = 1e-7

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model_path = r'./model'
train_info_path = r'./'
if not os.path.exists(model_path):
    os.mkdir(model_path)
```

## 数据集生成

```python
class Dataset:
    def __init__(self, domain):
        self.domain = domain
    
    def train_data(self, N=5000, verbose=None):
        # 内部点
        x_res = np.linspace(xmin, xmax, 100)
        y_res = np.linspace(ymin, ymax, 100)
        xx, yy = np.meshgrid(x_res, y_res)
        X_res = np.concatenate([xx.reshape((-1, 1)), yy.reshape((-1, 1))], axis=1)
        u_res = self.u_res_func(X_res)
        f_res = self.f_res_func(X_res)

        # 边界点
        observed_idx = np.concatenate([np.where(X_res[:,[0]] == xmin)[0], np.where(X_res[:,[0]] == xmax)[0], np.where(X_res[:,[1]] == ymin)[0], np.where(X_res[:,[1]] == ymax)[0]], axis=0)
        X_bcs = X_res[observed_idx]
        u_bcs = self.u_bcs_func(X_bcs)
        # T_data = temperature.reshape(-1, 1)[observed_idx]
        
        return X_res, X_bcs, u_bcs, u_res, f_res
    
    def u_bcs_func(self, X_res):
        f_bcs = (1/(2*(4*np.pi)**2)) * (np.sin(4*np.pi*X_res[:,[0]])) * (np.sin(4*np.pi*X_res[:,[1]])) 
        return f_bcs
    
    def u_res_func(self, X_res):
        f_res = (np.sin(4*np.pi*X_res[:,[0]])) * (np.sin(4*np.pi*X_res[:,[1]])) 
        return f_res
    
    def f_res_func(self, X_res):
        f_res = np.zeros((X_res.shape[0], 1))
        return f_res
    

dataset = Dataset(domain)
X_res, X_bcs, u_bcs, u_res, f_res = dataset.train_data()
print(X_res.shape, X_bcs.shape, f_res.shape)
X_res
u_res
```

## PINN主干网络

```python
class MLP(nn.Module):
    def __init__(self, mlp_layers):
        super(MLP, self).__init__()
        self.model = nn.Sequential()
        for i in range(len(mlp_layers)-2):
            self.model.add_module(f'fc{i+1}', nn.Linear(mlp_layers[i], mlp_layers[i+1], bias=True))
            self.model.add_module(f'act{i+1}', nn.Tanh())
        self.model.add_module(f'fc{len(mlp_layers)-1})', nn.Linear(mlp_layers[-2], mlp_layers[-1], bias=False))
    
    def forward(self, X):
        return self.model(X)
    
    
mlp = MLP(mlp_layers)
mlp_G = MLP(mlp_layers_G)
mlp.model
```

## PINN模型

```python
def grad(outputs, inputs):
    return torch.autograd.grad(outputs, inputs,
                               grad_outputs=torch.ones_like(outputs),
                               create_graph=True)
```

```python
class PINN(nn.Module):
    def __init__(self, backbone, nn_G, mu=None, sigma=None):
        super(PINN, self).__init__()
        self.backbone = backbone
        self.nn_G = nn_G
        
#         self.k = k
        if mu is not None and sigma is not None:
            self.is_inputs_normalization = True
            self.mu = mu
            self.sigma = sigma
            print(f'forward with normalization, mu={self.mu.tolist()}, sigma={self.sigma.tolist()}')
        else:
            self.is_inputs_normalization = False
    
    ## u_res_data
    def forward(self, X_res, X_bcs, u_res_data, u_bcs_data, f_res):
        f_res_pred = self.net_f(X_res)
#         u_res_pred = self.net_u(X_res)
        u_bcs_pred = self.net_u(X_bcs)
        
        penalty_weights = torch.ones(X_res.shape[0])
        # 取范围内的索引
        a = torch.where(X_res.abs()[:, 0] <= 0.01)[0]
        b = torch.where(X_res.abs()[:, 1] <= 0.01)[0]
        penalty_idx = a[torch.isin(a, b)]
        
        penalty_weights[penalty_idx] *= 10

        loss_res = torch.mean((penalty_weights*f_res_pred - f_res) ** 2)
#         loss_res_data = torch.mean((u_res_pred - u_res_data) ** 2)
        loss_bcs_data = torch.mean((u_bcs_pred - u_bcs_data)** 2)
        return loss_res,  loss_bcs_data
    
    def net_u(self, X):
        if self.is_inputs_normalization == True:
            X = (X - self.mu) / self.sigma
        return torch.tanh(self.backbone(X)) * 0.5 + 0.5
    
    def net_G(self, X):
        if self.is_inputs_normalization == True:
            X = (X - self.mu) / self.sigma
        return torch.tanh(self.nn_G(X)) * 0.5 + 0.5
    
    def net_f(self, X):
        X.requires_grad_(True)
        u_pred = self.net_u(X)
        
        # 求实部u的梯度
        grad_u = grad(u_pred, X)[0]
        u_x = grad_u[:, [0]]
        u_y = grad_u[:, [1]]
        
        u_xx = grad(u_x, X)[0][:, [0]]
        u_yy = grad(u_y, X)[0][:, [1]]
        
        G = self.net_G(X)
        
        f_res_pred = u_xx + u_yy + G
        
        return f_res_pred
    
    def penalty_loss_weight(X_res):
        penalty_weights = torch.ones(X_res.shape[0])
        print(penalty_weights)
        # 取范围内的索引
        a = torch.where(X_res.abs()[:, 0] <= 0.01)[0]
        b = torch.where(X_res.abs()[:, 1] <= 0.01)[0]
        penalty_idx = a[torch.isin(a, b)]

        penalty_weights[penalty_idx] *= 10
        return penalty_weights
    

pinn = PINN(mlp, mlp_G)
```

## Adam训练

```python
# 训练
dataset = Dataset(domain)
X_res, X_bcs, u_bcs, u_res, f_res = dataset.train_data()
X_res = torch.from_numpy(X_res).float().to(device)
X_bcs = torch.from_numpy(X_bcs).float().to(device)
u_bcs = torch.from_numpy(u_bcs).float().to(device)
u_res = torch.from_numpy(u_res).float().to(device)
f_res = torch.from_numpy(f_res).float().to(device)

mu = X_res.mean(dim=0)
sigma = X_res.std(dim=0)  # 求样本标准差

backbone = MLP(mlp_layers)
nn_G = MLP(mlp_layers_G)
pinn = PINN(backbone, nn_G, mu, sigma)
pinn = pinn.to(device)

optimizer_adam = optim.Adam(pinn.backbone.parameters(), lr=learning_rate)
optimizer_adam_param = optim.Adam(pinn.nn_G.parameters(), lr=learning_rate)
# optimizer, mode='min', factor=0.1, patience=10, threshold=0.0001, 
# threshold_mode='rel', cooldown=0, min_lr=0, eps=1e-08, verbose=False
lr_sche = ReduceLROnPlateau(optimizer_adam, mode='min', factor=gamma, verbose=True)
lr_sche_param = ReduceLROnPlateau(optimizer_adam_param, mode='min', factor=gamma, verbose=True)

logger = {
    "loss": [],
    "loss_res": [],
    "loss_res_data": [],
    "loss_bcs_data": [],
    "iter": [],
    "mu": mu,
    "sigma": sigma
}
best_loss = 1e9
with open(train_info_path + 'train_info.txt', 'w') as f:
    f.write('Training by Adam:\n')
start_time = time.time()

for it in range(n_iters):
    pinn.zero_grad()
#     loss_res, loss_data = pinn(X_res, X_bcs, u_res, u_bcs)
#     loss = 10*loss_res + 1000*loss_data
    loss_res, loss_bcs_data = pinn(X_res, X_bcs, u_res, u_bcs, f_res)
    loss = loss_res + 1000*(loss_bcs_data)
#     loss = 100*loss_res + loss_bcs_data
    loss.backward()
    optimizer_adam.step()
    optimizer_adam_param.step()
    
    # 保存训练信息 -------
    logger["loss"].append(loss.item())
    logger["loss_res"].append(loss_res.item())
#     logger["loss_res_data"].append(loss_res_data.item())
    logger["loss_bcs_data"].append(loss_bcs_data.item())
    logger["iter"].append(it + 1)
    
    if (it + 1) % 100 == 0:  # 打印训练信息 -------
        info = f'Iter # {it+1:4d}/{n_iters}\ttime:{time.time()-start_time:.1f}\t' + f'loss:{loss.item():.2e}, loss_r:{loss_res.item():.2e}, loss_b_d:{loss_bcs_data.item():.2e}'
        with open(train_info_path + 'train_info.txt', 'a') as f:
            f.write(info + '\n')
        print(info)
        # 保存训练loss最低的模型
        if loss.item() < best_loss:
            model_state = {'iter': it+1, 'backbone_state': pinn.backbone.state_dict(), 'nn_G_state': pinn.nn_G.state_dict()}
            torch.save(model_state, os.path.join(model_path, 'pinn_adam.pth'))
            best_loss = loss.item()
        
    if (it + 1) % 100 == 0:  # 学习率衰减 -------
        lr_sche.step(loss)
        lr_sche_param.step(loss)
```
