```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
```

```python
x = np.linspace(0, 0.25, 101)
y = np.linspace(0, 0.25, 101)
xx, yy = np.meshgrid(x, y)
xx = xx.flatten()
yy = yy.flatten()

G = np.sin(4 * np.pi * xx) * np.sin(4 * np.pi * yy)
T = 1/(2 * (4 * np.pi)**2) * np.sin(4 * np.pi * xx) * np.sin(4 * np.pi * yy)

xx = xx.reshape((101, 101))
yy = yy.reshape((101, 101))
G = G.reshape((101, 101))
T = T.reshape((101, 101))
```

```python
plt.pcolor(xx, yy, G)
plt.colorbar()
```

```python
plt.pcolor(xx, yy, T)
plt.colorbar()
```

```python
data = {'水平方向': xx.flatten(), '垂直方向': yy.flatten(), '温度': T.flatten(), '热源': G.flatten()}

df = pd.DataFrame(data)  
df.to_excel('./温度和热源.xlsx', index=False)
```
