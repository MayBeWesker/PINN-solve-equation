# 网络

```python
class MLP(nn.Module):
    def __init__(self, mlp_layers):
        super(MLP, self).__init__()
        
        self.model = nn.Sequential()
        for i in range(len(mlp_layers)-2):
            layer = nn.Sequential()
            layer.add_module(f'fc{i}', nn.Linear(mlp_layers[i], mlp_layers[i+1], bias=True))
            layer.add_module(f'act{i}', nn.Tanh())
            self.model.add_module(f'layer{i}', layer)

        last_layer = nn.Sequential()
        last_layer.add_module(f'fc{len(mlp_layers)-2}', nn.Linear(mlp_layers[-2], mlp_layers[-1], bias=False))
        self.model.add_module(f'layer{len(mlp_layers)-2}', last_layer)
        
#         for param in self.parameters():
#             if len(param.shape) > 1:
#                 nn.init.kaiming_normal_(param)
    
    def forward(self, X):
        return self.model(X)

    
model_psi = MLP(layers)
model_p = MLP(layers)
model_phil = MLP(layers)
# model
```

# Resample

```python
# net_fi表示net_u, net_v, ......
def prob_resample(X_resami, net_fi, num=100, eps=0.01, plot=True):
    # 获得residule
    fi = net_fi(X_resami)
    fi = fi.detach().cpu().numpy()
    # 计算概率权重
    prob = np.concatenate([np.log(abs(fi / eps) + np.finfo(np.float32).eps), np.zeros(fi.shape)], axis=1)
    print(prob)
    prob = np.max(prob, axis=1)
    prob = prob / sum(prob)
    # np.random.choice得到索引
    idx = np.random.choice(X_resami.shape[0], num, replace=True, p=prob)
    return X_resami[idx].detach()

def easy_resample(X_res, net_f, device=torch.device('cuda'), n_total_points=4000, n_added_points=100):
    lb, ub = get_lb_ub(X_res)
    # 生成待采样的点集
    X_resam = (ub - lb)*lhs(3,n_total_points) + lb
    X_resam = X_resam.float().to(device)
    # 获得residule
    f = net_f(X_resam)
    f = f.detach().cpu().numpy()
    # 依loss_res的值排序得到索引
    idx = np.argsort(abs(f).flatten())[-n_added_points:]
    return X_resam[idx].detach()
```

# PINN

```python
def grad(outputs, inputs):
    return torch.autograd.grad(outputs, inputs,
                               grad_outputs=torch.ones_like(outputs),
                               create_graph=True)
```

```python
class PINN(nn.Module):
    def __init__(self, model_psi, model_p, model_phil, mu=None, sigma=None):
        super(PINN, self).__init__()
        self.model_psi = model_psi
        self.model_p = model_p
        self.model_phil = model_phil
        
        if mu is not None and sigma is not None:
            self.is_inputs_normalization = True
            self.mu = mu
            self.sigma = sigma
            print(f'forward with normalization, mu={self.mu.tolist()}, sigma={self.sigma.tolist()}')
        else:
            self.is_inputs_normalization = False
    
    def forward(self, X_res_psi, X_res_p, X_res_phil,  u_data, v_data, p_data, phil_data):
        u_pred, v_pred = self.net_psi(X_res_psi)
        loss_u = torch.mean((u_pred-u_data)**2)
        loss_v = torch.mean((v_pred-v_data)**2)
        loss_p = torch.mean((self.net_p(X_res_p)-p_data)**2)
        loss_phil = torch.mean((self.net_phil(X_res_phil)-phil_data)**2)
        loss = loss_u + loss_v + loss_p + loss_phil
        return loss, loss_u, loss_v, loss_p, loss_phil
    
    def net_psi(self, X):
        if self.is_inputs_normalization == True:
            X = (X - self.mu) / self.sigma
        X.requires_grad_(True)
        psi = self.model_psi(X)
        
        grad_psi = grad(psi, X)[0]
        
        u_pred = grad_psi[:,[1]]
        v_pred = -grad_psi[:,[0]]

        return u_pred, v_pred
    
    def net_p(self, X):
        if self.is_inputs_normalization == True:
            X = (X - self.mu) / self.sigma
        return self.model_p(X)
    
    def net_phil(self, X):
        if self.is_inputs_normalization == True:
            X = (X - self.mu) / self.sigma
        return self.model_phil(X)
    

pinn = PINN(model_psi, model_p, model_phil)
```

```python
x_train = torch.from_numpy(x_train).float().to(device)
y_train = torch.from_numpy(v_train).float().to(device)
t_train = torch.from_numpy(t_train).float().to(device)
u_train = torch.from_numpy(u_train).float().to(device)
v_train = torch.from_numpy(v_train).float().to(device)
p_train = torch.from_numpy(p_train).float().to(device)
phil_train = torch.from_numpy(phil_train).float().to(device)

X_train_psi = torch.from_numpy(X_train_psi).float().to(device)
X_train_p = torch.from_numpy(X_train_p).float().to(device)
X_train_phil = torch.from_numpy(X_train_phil).float().to(device)
X_psi = torch.from_numpy(X_psi).float().to(device)
X_p = torch.from_numpy(X_p).float().to(device)
X_phil = torch.from_numpy(X_phil).float().to(device)


# mu = X_train_u.mean(dim=0)
# sigma = X_train_u.std(dim=0)  # 求样本标准差

model_psi = MLP(layers)
model_p = MLP(layers)
model_phil = MLP(layers)
pinn = PINN(model_psi, model_p, model_phil).to(device)

optimizer_adam = optim.Adam(pinn.parameters(), lr=1e-3)

lr_sche = optim.lr_scheduler.ExponentialLR(optimizer_adam, gamma=0.9)  # 指数衰减学习率

logger_adam = {
    "loss": [], 
    "loss_u": [],
    "loss_v": [],
    "loss_p": [],
    "loss_phil": [],
    "iter": [],
}
best_loss = 1e9

# 训练
start_time = time.time()
for it in range(adam_iters):
    pinn.train()
    pinn.zero_grad()
    
    loss, loss_u, loss_v, loss_p, loss_phil = pinn(X_train_psi, X_train_p, X_train_phil, u_train, v_train, p_train, phil_train)

    loss.backward()
    optimizer_adam.step()
    
    if (it + 1) % 100 == 0:
        # 保存loss信息
        pinn.train(False)
        loss_valid, loss_u_valid, loss_v_valid, loss_p_valid, loss_phil_valid = pinn(X_train_psi, X_train_p, X_train_phil,  u_train, v_train, p_train, phil_train)
        
        logger_adam["loss"].append(loss_valid.item())
        logger_adam["loss_u"].append(loss_u_valid.item())
        logger_adam["loss_v"].append(loss_v_valid.item())
        logger_adam["loss_p"].append(loss_p_valid.item())
        logger_adam["loss_phil"].append(loss_phil_valid.item())
        logger_adam["iter"].append(it+1)
        
        # 保存训练loss最低的模型
        if loss_valid.item() < best_loss:
            model_state = {'iter': it+1, 'model_psi_state': pinn.model_psi.state_dict(), 'model_p_state': pinn.model_p.state_dict(), 'model_phil_state': pinn.model_phil.state_dict()}
            torch.save(model_state, os.path.join(model_path, 'pinn_adam.pth'))
            best_loss = loss_valid.item()
        
        if (it + 1) % 500 == 0:
            # 保存并打印训练日志
            info = f'Iter # {it+1:6d}/{adam_iters}\t' + \
                f'loss:{loss.item():.2e}, loss_u:{loss_u.item():.2e}, loss_v:{loss_v.item():.2e}, loss_p:{loss_p.item():.2e}, loss_phil:{loss_phil.item():.2e},   ' + \
                f'Valid # loss:{loss_valid.item():.2e}, loss_u:{loss_u_valid.item():.2e}, loss_v:{loss_v_valid.item():.2e}, loss_p:{loss_p_valid.item():.2e}, loss_phil:{loss_phil_valid.item():.2e}'
            with open(train_info_path + 'train_info.txt', 'a') as f:
                f.write(info + '\n')
            print(info)
            
        if (it + 1) % 4000 == 0:
            lr_sche.step()
            
#     if (it +1) % 10 == 0:
#         pinn.zero_grad()
#         pinn.eval()
#         # 进行重采样
# #         print(X_res.shape)
#         X_resam_u = easy_resample(X_u, pinn.net_u).float()
#         X_resam_v = easy_resample(X_v, pinn.net_u).float()
#         X_resam_p = easy_resample(X_p, pinn.net_u).float()
#         X_resam_phil = easy_resample(X_phil, pinn.net_u).float()
#         # 拼接数据
#         X_train_u = torch.cat([X_train_u, X_resam_u], dim=0)
#         X_train_v = torch.cat([X_train_v, X_resam_u], dim=0)
#         X_train_p = torch.cat([X_train_p, X_resam_u], dim=0)
#         X_train_phil = torch.cat([X_train_phil, X_resam_u], dim=0)
#         print(X_train_u.shape)
#         # 拼接权重
#         print(w_u.shape)
#         w_u_resam = torch.full((X_resam_u.shape[0],1),torch.mean(w_u).item()).cuda()
#         w_u = torch.cat([w_u,w_u_resam], dim=0)
#         print(w_u.shape)
```
